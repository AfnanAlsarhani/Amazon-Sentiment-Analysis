## Step 1: Load the Dataset

We load a sample of Amazon product reviews containing text and star ratings.

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("amazon_reviews_sample.csv")

# Display the first few rows
df.head()
```

```python
# Install TextBlob if not already installed
!pip install -q textblob

# Import necessary tools
from textblob import TextBlob
```

## Step 2: Sentiment Classification

Using TextBlob, we calculate the polarity of each review and classify it into:  
- **Positive** if polarity > 0.1  
- **Negative** if polarity < -0.1  
- **Neutral** if in between


```python
# Function to classify sentiment based on polarity score
def get_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity
    if polarity > 0.1:
        return 'Positive'
    elif polarity < -0.1:
        return 'Negative'
    else:
        return 'Neutral'

# Apply sentiment analysis to each review
df['sentiment'] = df['review'].apply(get_sentiment)

# Show updated dataset with sentiment
df[['review', 'rating', 'sentiment']]
```

## Step 3: Visualize Sentiment Distribution

We plot a pie chart to show the percentage of positive, neutral, and negative reviews.


```python
import matplotlib.pyplot as plt

# Count of each sentiment
sentiment_counts = df['sentiment'].value_counts()

# Plot a pie chart
plt.figure(figsize=(6, 6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', colors=['lightgreen', 'gold', 'salmon'])
plt.title('Sentiment Distribution in Amazon Reviews')
plt.show()
```
